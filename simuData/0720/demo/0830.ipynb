```python
from sympy import *

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure(figsize=(10, 5), dpi=300)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
x_cal = np.logspace(1, 10, 10**6, base=10)
y_calG = lambdify(omega, 20*log(abs(Av.subs({s: I*omega})), 10), "numpy")(x_cal)
y_calP = lambdify(omega, deg(arg(Av.subs({s: I*omega}))), "numpy")(x_cal)
x_rG,y_rG = np.loadtxt('realGain.csv', delimiter=',',unpack=True)
x_sG,y_sG = np.loadtxt('smallGain.csv', delimiter=',',unpack=True)
x_rP,y_rP = np.loadtxt('realPhase.csv', delimiter=',',unpack=True)
x_sP,y_sP = np.loadtxt('smallPhase.csv', delimiter=',',unpack=True)
ax1.set_xscale("log")
ax1.set_ylabel(r"$20{\rm lg} |A_v|$")
ax1.plot(x_cal, y_calG, label = "calGain")
ax1.plot(x_rG, y_rG, label = "realGain")
ax1.plot(x_sG, y_sG, label = "smallGain")
ax1.legend()
ax1.grid()
ax2.set_xlabel(r"$\omega\,\rm / Hz$")
ax2.set_ylabel(r"$\arg A_v\,\rm / deg$")
ax2.set_xscale("log")
ax2.plot(x_cal, y_calP,  label = "calPhase")
ax2.plot(x_rP, y_rP-180, label = "realPhase")
ax2.plot(x_sP, y_sP, label = "smallPhase")
ax2.legend()
ax2.grid()
```
