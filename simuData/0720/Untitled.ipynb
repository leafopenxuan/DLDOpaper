```python
from sympy import *

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure(figsize=(8, 6), dpi=300)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
t0, out0 = np.loadtxt('out_0volt.csv', delimiter=',',unpack=True)
t1, out1 = np.loadtxt('out_1volt.csv', delimiter=',',unpack=True)
t, output = np.loadtxt('outputvolt.csv', delimiter=',',unpack=True)
ax1.plot(t*10**6, output, label = "output", alpha=0.6, color="red")
ax1.plot(t0*10**6, out0, label = "output 0", color="green")
ax1.legend()
ax1.grid()
ax2.plot(t*10**6, output, label = "output", alpha=0.6, color="red")
ax2.plot(t1*10**6, out1, label = "output 1", color="blue")
ax2.legend()
ax2.grid()
ax2.set_xlabel(r"$\rm\mu S$")
ax1.set_ylabel(r"v")
ax2.set_ylabel(r"v")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=300)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
t0, out0 = np.loadtxt('out_0Curr.csv', delimiter=',',unpack=True)
t1, out1 = np.loadtxt('out_1Curr.csv', delimiter=',',unpack=True)
# t, output = np.loadtxt('outputvolt.csv', delimiter=',',unpack=True)
# ax1.plot(t*10**6, output, label = "output", alpha=0.6, color="red")
ax1.plot(t0*10**6, 1000*out0, label = "load current 0", color="green")
ax1.legend(loc=4)
ax1.grid()
# ax2.plot(t*10**6, output, label = "output", alpha=0.6, color="red")
ax2.plot(t1*10**6, 1000*out1, label = "load current 1", color="blue")
ax2.legend(loc=4)
ax2.grid()
ax2.set_xlabel(r"$\rm\mu S$")
ax1.set_ylabel(r"mA")
ax2.set_ylabel(r"mA")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=300)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
t0, out0 = np.loadtxt('gnd_0Volt.csv', delimiter=',',unpack=True)
t1, out1 = np.loadtxt('gnd_1Volt.csv', delimiter=',',unpack=True)
t, output = np.loadtxt('gndVolt.csv', delimiter=',',unpack=True)
ax1.plot(t0*10**6, output, label = r"$\rm V_{SSB}$ at shared AA loop", alpha=0.6, color="red")
ax1.plot(t*10**6, out0, label = "Analog ground at lat_0", color="green")
ax1.legend(loc=4)
ax1.grid()
ax2.plot(t1*10**6, output, label = r"$\rm V_{SSB}$ at shared AA loop", alpha=0.6, color="red")
ax2.plot(t*10**6, out1, label = "Analog ground at lat_1", color="blue")
ax2.legend(loc=4)
ax2.grid()
ax2.set_xlabel(r"$\rm\mu S$")
ax1.set_ylabel(r"V")
ax2.set_ylabel(r"V")
```
